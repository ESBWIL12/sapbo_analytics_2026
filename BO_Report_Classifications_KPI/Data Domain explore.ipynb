```python
%sql
SELECT DISTINCT 
  e.Document_name,
  e.entry_type,
  e.entry_name, 
  e.variable_definition,
  e.referenced_datafield_name,
  e.datafield_description, 
  e.Universe_Name,
  e.sql_definition,
  e.UNIVERSE_PRIMARY_CATEGORY,
  l.BO_TABLE,
  l.BO_SCHEMA,
  l.Calc_source_table,
  l.Calc_source_schema,
  l.Databricks_Schema,
  l.databricks_ingested
FROM dataplatform01_central_dev_catalog_01.custom_sap_bo.webi_data_entries e
LEFT JOIN dataplatform01_central_dev_catalog_01.custom_sap_bo.active_webi_dataentry_linage_ucflagged l
  ON e.Document_Id = l.Document_Id
  AND (
    LOWER(e.sql_definition) LIKE CONCAT('%', LOWER(l.BO_TABLE), '%')
    OR LOWER(e.datafield_datasourcepath) LIKE CONCAT('%', LOWER(l.BO_TABLE), '%')
  )
WHERE 
  -- Search entry_name for claims received patterns
  (
    LOWER(e.entry_name) LIKE '%claim%receiv%'
    OR LOWER(e.entry_name) LIKE '%receiv%claim%'
    OR LOWER(e.entry_name) LIKE '%nb%claim%'
    OR LOWER(e.entry_name) LIKE '%#%claim%'
    OR LOWER(e.entry_name) LIKE '%number%claim%'
    OR LOWER(e.entry_name) LIKE '%count%claim%'
  )
  -- Or search variable_definition / datafield_description
  OR (
    LOWER(e.variable_definition) LIKE '%claim%' 
    AND LOWER(e.variable_definition) LIKE '%receiv%'
    AND (
      LOWER(e.variable_definition) LIKE '%count%' 
      OR LOWER(e.variable_definition) LIKE '%number%'
      OR LOWER(e.variable_definition) LIKE '%unique%'
      OR LOWER(e.variable_definition) LIKE '%filed%'
      OR LOWER(e.variable_definition) LIKE '%registered%'
      OR LOWER(e.variable_definition) LIKE '%period%'
    )
  )
  OR (
    LOWER(e.datafield_description) LIKE '%claim%' 
    AND LOWER(e.datafield_description) LIKE '%receiv%'
    AND (
      LOWER(e.datafield_description) LIKE '%count%' 
      OR LOWER(e.datafield_description) LIKE '%number%'
      OR LOWER(e.datafield_description) LIKE '%unique%'
      OR LOWER(e.datafield_description) LIKE '%filed%'
      OR LOWER(e.datafield_description) LIKE '%registered%'
      OR LOWER(e.datafield_description) LIKE '%period%'
    )
  )
ORDER BY e.entry_name
```

```python
%sql
SELECT DISTINCT 
  e.Document_name,
  e.entry_type,
  e.entry_name, 
  e.variable_definition,
  e.referenced_datafield_name,
  e.datafield_description, 
  e.Universe_Name,
  e.sql_definition,
  e.UNIVERSE_PRIMARY_CATEGORY,
  l.BO_TABLE,
  l.BO_SCHEMA,
  l.Calc_source_table,
  l.Calc_source_schema,
  l.Databricks_Schema,
  l.databricks_ingested
FROM dataplatform01_central_dev_catalog_01.custom_sap_bo.webi_data_entries e
LEFT JOIN dataplatform01_central_dev_catalog_01.custom_sap_bo.active_webi_dataentry_linage_ucflagged l
  ON e.Document_Id = l.Document_Id
  AND (
    LOWER(e.sql_definition) LIKE CONCAT('%', LOWER(l.BO_TABLE), '%')
    OR LOWER(e.datafield_datasourcepath) LIKE CONCAT('%', LOWER(l.BO_TABLE), '%')
  )
WHERE 
  -- Search entry_name for claims paid patterns
  (
    LOWER(e.entry_name) LIKE '%claim%paid%'
    OR LOWER(e.entry_name) LIKE '%paid%claim%'
    OR LOWER(e.entry_name) LIKE '%claim%payment%'
    OR LOWER(e.entry_name) LIKE '%payment%claim%'
    OR LOWER(e.entry_name) LIKE '%claim%amount%'
    OR LOWER(e.entry_name) LIKE '%amount%paid%'
    OR LOWER(e.entry_name) LIKE '%claim%settled%'
  )
  -- Or search variable_definition for claims paid context
  OR (
    LOWER(e.variable_definition) LIKE '%claim%' 
    AND (
      LOWER(e.variable_definition) LIKE '%paid%'
      OR LOWER(e.variable_definition) LIKE '%payment%'
      OR LOWER(e.variable_definition) LIKE '%settled%'
    )
    AND (
      LOWER(e.variable_definition) LIKE '%amount%' 
      OR LOWER(e.variable_definition) LIKE '%total%'
      OR LOWER(e.variable_definition) LIKE '%sum%'
      OR LOWER(e.variable_definition) LIKE '%closed%'
      OR LOWER(e.variable_definition) LIKE '%period%'
    )
  )
  -- Or search datafield_description for claims paid context
  OR (
    LOWER(e.datafield_description) LIKE '%claim%' 
    AND (
      LOWER(e.datafield_description) LIKE '%paid%'
      OR LOWER(e.datafield_description) LIKE '%payment%'
      OR LOWER(e.datafield_description) LIKE '%settled%'
    )
    AND (
      LOWER(e.datafield_description) LIKE '%amount%' 
      OR LOWER(e.datafield_description) LIKE '%total%'
      OR LOWER(e.datafield_description) LIKE '%sum%'
      OR LOWER(e.datafield_description) LIKE '%closed%'
      OR LOWER(e.datafield_description) LIKE '%period%'
      OR LOWER(e.datafield_description) LIKE '%financial%'
    )
  )
ORDER BY e.entry_name
```

```python
%sql
SELECT DISTINCT 
  e.Document_name,
  e.entry_type,
  e.entry_name, 
  e.variable_definition,
  e.referenced_datafield_name,
  e.datafield_description, 
  e.Universe_Name,
  e.sql_definition,
  e.UNIVERSE_PRIMARY_CATEGORY,
  l.BO_TABLE,
  l.BO_SCHEMA,
  l.Calc_source_table,
  l.Calc_source_schema,
  l.Databricks_Schema,
  l.databricks_ingested
FROM dataplatform01_central_dev_catalog_01.custom_sap_bo.webi_data_entries e
LEFT JOIN dataplatform01_central_dev_catalog_01.custom_sap_bo.active_webi_dataentry_linage_ucflagged l
  ON e.Document_Id = l.Document_Id
  AND (
    LOWER(e.sql_definition) LIKE CONCAT('%', LOWER(l.BO_TABLE), '%')
    OR LOWER(e.datafield_datasourcepath) LIKE CONCAT('%', LOWER(l.BO_TABLE), '%')
  )
WHERE 
  -- Search entry_name for NNP / non-payment notification patterns
  (
    LOWER(e.entry_name) LIKE '%nnp%'
    OR LOWER(e.entry_name) LIKE '%non-payment%'
    OR LOWER(e.entry_name) LIKE '%non payment%'
    OR LOWER(e.entry_name) LIKE '%nonpayment%'
    OR LOWER(e.entry_name) LIKE '%notification%non%pay%'
    OR LOWER(e.entry_name) LIKE '%nb%nnp%'
    OR LOWER(e.entry_name) LIKE '%#%nnp%'
    OR LOWER(e.entry_name) LIKE '%number%nnp%'
    OR LOWER(e.entry_name) LIKE '%count%nnp%'
  )
  -- Or search variable_definition for NNP context
  OR (
    LOWER(e.variable_definition) LIKE '%claim%' 
    AND (
      LOWER(e.variable_definition) LIKE '%nnp%'
      OR LOWER(e.variable_definition) LIKE '%non-payment%'
      OR LOWER(e.variable_definition) LIKE '%non payment%'
      OR LOWER(e.variable_definition) LIKE '%nonpayment%'
    )
    AND (
      LOWER(e.variable_definition) LIKE '%notification%' 
      OR LOWER(e.variable_definition) LIKE '%number%'
      OR LOWER(e.variable_definition) LIKE '%count%'
      OR LOWER(e.variable_definition) LIKE '%total%'
      OR LOWER(e.variable_definition) LIKE '%unpaid%'
      OR LOWER(e.variable_definition) LIKE '%period%'
    )
  )
  -- Or search datafield_description for NNP context
  OR (
    LOWER(e.datafield_description) LIKE '%claim%' 
    AND (
      LOWER(e.datafield_description) LIKE '%nnp%'
      OR LOWER(e.datafield_description) LIKE '%non-payment%'
      OR LOWER(e.datafield_description) LIKE '%non payment%'
      OR LOWER(e.datafield_description) LIKE '%nonpayment%'
    )
    AND (
      LOWER(e.datafield_description) LIKE '%notification%' 
      OR LOWER(e.datafield_description) LIKE '%number%'
      OR LOWER(e.datafield_description) LIKE '%count%'
      OR LOWER(e.datafield_description) LIKE '%total%'
      OR LOWER(e.datafield_description) LIKE '%unpaid%'
      OR LOWER(e.datafield_description) LIKE '%period%'
    )
  )
ORDER BY e.entry_name
```
